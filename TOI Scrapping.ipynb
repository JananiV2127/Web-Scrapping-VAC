```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = "https://timesofindia.indiatimes.com/"
```

```python
allow = requests.get(url)
```

```python
print(allow)
```

```python
soup = BeautifulSoup(allow.text)
```

```python
soup
```

```python
ipl_2022 = soup.select("._9_syI a")
```

```python
ipl_2022
```

```python
ipl_titles = []
ipl_link = []
```

```python
for item in ipl_2022:
    ipl_titles.append(item.text)
    link = "https://timesofindia.indiatimes.com/" + item["href"]
    ipl_link.append(link)
```

```python
ipl_titles
```

```python
ipl_link
```

```python
import pandas as pd
```

```python
toi = pd.DataFrame(ipl_titles,columns=['Titles'])
```

```python
toi
```

```python
toi.to_csv('toi.csv')
```
